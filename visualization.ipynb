```python
import os 
import matplotlib.pyplot as plt 
```

```python
dir_path = os.path.abspath('')
dataset_folder = os.path.join(dir_path,'Dataset')
training_folder = os.path.join(dataset_folder,'training')
validation_folder = os.path.join(dataset_folder,'validation')
evaluation_folder = os.path.join(dataset_folder,'evaluation')
```

```python
training_food_folder = os.path.join(training_folder,'Food')
training_not_food_folder = os.path.join(training_folder,'Not Food')

validation_food_folder = os.path.join(validation_folder,'Food')
validation_not_food_folder = os.path.join(validation_folder,'Not Food')

evaluation_food_folder = os.path.join(evaluation_folder,'Food')
evaluation_not_food_folder = os.path.join(evaluation_folder,'Not Food')
```

```python
print('Number of training examples:',len(os.listdir(training_food_folder))+len(os.listdir(training_not_food_folder)))
print('Number of validation examples:',len(os.listdir(validation_food_folder))+len(os.listdir(validation_not_food_folder)))
print('Number of evaluation examples:',len(os.listdir(evaluation_food_folder))+len(os.listdir(evaluation_not_food_folder)))
```

```python
train_food_fname = os.listdir(training_food_folder)
train_not_food_fname = os.listdir(training_not_food_folder)

print(train_food_fname[:10])
print(train_not_food_fname[:10])
```

```python
%matplotlib inline

import matplotlib.image as mpimg

nrows = 4
ncols = 4

pic_index = 0
```

```python
fig = plt.gcf()
fig.set_size_inches(ncols*4, nrows*4)

pic_index+=8

next_food_pix = [os.path.join(training_food_folder, fname) 
                for fname in train_food_fname[ pic_index-8:pic_index] 
               ]

next_not_food_pix = [os.path.join(training_not_food_folder, fname) 
                for fname in train_not_food_fname[ pic_index-8:pic_index]
               ]

for i, img_path in enumerate(next_food_pix+next_not_food_pix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(nrows, ncols, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```
